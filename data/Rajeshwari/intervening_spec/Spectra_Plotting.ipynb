```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import json

data = np.sort( glob('spectra/*') )
plt.figure(figsize=(15, 24))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Intervening Spectra", fontsize=18, y=0.95)

for n, i in enumerate(data):
    ax = plt.subplot(7, 2, n + 1)
    
    name = i.split('/')[1].split('SPEC_')[1].split('.')[0]
    
    loaded = np.loadtxt(i)

    ax.plot(loaded[:,0], loaded[:,1])

    ax.set_title(name)
    ax.set_xlabel("heliocentric frequency (MHz)")
    ax.set_ylabel('continuum-subtracted flux (mJy)')
```
