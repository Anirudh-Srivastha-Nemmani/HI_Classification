```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
path = np.sort(glob('./21cm_spc/*'))
```

```python
len(path)
```

```python
plt.figure(figsize=(15, 30))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Intervening Spectra", fontsize=18)
for n, i in enumerate(path):
    ax = plt.subplot(16, 2, n + 1)
    name = i.split('/')[-1]
    if name == '0237-233_gbt.freq':
        loaded = np.loadtxt(i)
        ax.plot(loaded[:,0], loaded[:,1])
        ax.set_title(name)
        fig = ax.get_figure()
        fig.tight_layout()
        fig.subplots_adjust(top=0.95)
    else:
        loaded = np.loadtxt(i)
        ax.plot(loaded[:,1], loaded[:,2])
        ax.set_title(name)
        fig = ax.get_figure()
        fig.tight_layout()
        fig.subplots_adjust(top=0.95)
```

```python
gg = './21cm_spc/1621+074_gmrt.freq'#path[30]
name = gg.split('/')[-1]
loaded = np.loadtxt(gg)
plt.plot(loaded[:,2], loaded[:,3])
plt.title(name)
plt.show()
```
