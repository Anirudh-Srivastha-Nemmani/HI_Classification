```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import json

data = np.sort( glob('Intervening_reversed/*') )
plt.figure(figsize=(15, 30))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Intervening Spectra", fontsize=18)
print(len(data))
for n, i in enumerate(data):
    ax = plt.subplot(15, 2, n + 1)
    
    name = i.split('/')[1]
    
    loaded = np.loadtxt(i)

    ax.plot(loaded[:,0], loaded[:,1])

    ax.set_title(name)
    ax.set_xlabel("heliocentric frequency (MHz)")
    ax.set_ylabel('continuum-subtracted flux (mJy)')
    fig = ax.get_figure()
    fig.tight_layout()
    fig.subplots_adjust(top=0.95)
```
