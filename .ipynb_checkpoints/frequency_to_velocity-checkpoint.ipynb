# Intervening

```python
i = './data/Kanekar/Spectra_flux_vs_frequency/0237-233_gbt.freq'
name = i.split('/')[-1]
z0 = 1.6724
f0 = 1420.405752/(1 + z0)
loaded = np.loadtxt(i)
c = 299792.458
velocity = [((f0 - x)*c)/x for x in loaded[:,0]]
velocity = np.array(velocity)
loaded[:,0] = velocity
loaded[:,1]*= -1
np.savetxt('./Intervening_velocity_spectrum/'+name, loaded)

plt.plot(loaded[:,0], loaded[:,1])
plt.title(name)
plt.show()
```
