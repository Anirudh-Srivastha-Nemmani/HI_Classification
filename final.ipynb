```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
list_ = np.sort(glob('./Intervening_velocity_spectrum/*'))
print(len(list_))
```

```python
plt.figure(figsize=(15, 60))
plt.subplots_adjust(hspace=0.5)
plt.suptitle("Intervening Spectra", fontsize=18)
for n, i in enumerate(list_):
    ax = plt.subplot(23, 2, n + 1)
    name = i.split('/')[-1]
    loaded = np.loadtxt(i)
    ax.plot(loaded[:,0], loaded[:,1])
    ax.set_title(name)
    fig = ax.get_figure()
    fig.tight_layout()
    fig.subplots_adjust(top=0.95)
```
